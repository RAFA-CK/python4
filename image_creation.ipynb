```python
# Read image using OpenCV
image = cv2.imread(image_path)  # Read in BGR format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Display the original image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Convert to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display both images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title("RGB Image")
ax[0].axis("off")

ax[1].imshow(image_gray, cmap="gray")
ax[1].set_title("Grayscale Image")
ax[1].axis("off")

plt.show()

# Print pixel values
print("Pixel Value at (50,50) in RGB:", image_rgb[50, 50])
print("Pixel Value at (50,50) in Grayscale:", image_gray[50, 50])
```

```python
# Resize the image (scale down by 50%)
scaled_image = cv2.resize(image_rgb, None, fx=0.5, fy=0.5)

# Display
plt.imshow(scaled_image)
plt.title("LION")
plt.axis("off")
plt.show()
```

```python
# Get image center and define rotation matrix
(h, w) = image_rgb.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 30, 1)  # Rotate by 45 degrees

# Apply rotation
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (w, h))

# Display rotated image
plt.imshow(rotated_image)
plt.title("Rotated Image (30)")
plt.axis("off")
plt.show()
```
